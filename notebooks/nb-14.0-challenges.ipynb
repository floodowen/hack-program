### Challenge 1: 
Execute the code cell below to see an example of how it works. 
Use markdown in the cell after the code-block to describe the function `random_words_api`. Try to be descriptive about what each step of code in this function does, and why it works. 

```python
def random_words_api(nwords=10):
    "no docstring"
    URL = "https://random-word-api.herokuapp.com/word"
    response = requests.get(url=URL, params={"number": nwords})
    return response.json()

# demonstration
random_words_api(5)
```

**Description:** The function `random_words_api` will obtain a list of words from a heroku wep app. The argument _nwords_ will define how many words are returned in this list. The _URL_ variable tells the function which website the app uses. The _response_ variable will save the request output, which uses the url that was defined as _URL_, and the paramaters of the number equally the previously defined _nwords_. The return line will return the saved _response_ variable in JSON format.

### Challenge 2: 
Use the `random_words_api` function to get 50 random words and store the result as a variable. Write a function that takes the list of words as input and returns a dictionary with the longest word as the key and the length of the longest word as a value. If there is a tie in the length of words then have it return additional words as keys with their lengths as values.

```python
stored = random_words_api(50)

def longest_word(stored):
    longsorted = sorted(stored,key=len)
    output = {}
    for word in longsorted:
        if len(word) == len(longsorted[-1]):
            output[word] = len(word)
    return output
print(longest_word(stored))
```

### Challenge 3: 
Write a function to take the list of words as input and trim all words to be at most 5 characters in length, and return as a list.

```python
def word_trim(stored):
    trimmed_words = []
    for word in stored:
        trimmed_words.append(word[:5])
    return trimmed_words
print(word_trim(stored))
```

### Challenge 4: 
Write a function to take a list of words as input and to count the occurrence of all letters in every word and return as a dictionary mapping letters to integers., e.g., {'a': 10, 'b': 3, 'c': 5, ...}.  

```python
def occurence_count(stored):
    concat = "".join(stored)
    alphabet_count = {}
    for letters in concat:
        if letters not in alphabet_count:
            alphabet_count[letters] = 1
        else:
            alphabet_count[letters] += 1
    return alphabet_count
result = occurence_count(stored)
for key in sorted(result.keys()):
    print(key, result[key])
```
